```python
import numpy as np
import pandas as pd
```

```python
# tmp_A = np.random.randint(10, size=(5, 4))
# tmp_p = np.random.rand(4)

tmp_A = np.array([[6, 9, 3],
         [5, 10, 30],
         [8, 12, 4]])

tmp_p = np.array([0.3, 0.4, 0.3])
tmp_lamb = 0.8
```

# Критерий Вальда

```python
def wald(A):
    v = np.min(A, axis=1)
    return np.argmax(v) + 1
```

```python
wald(tmp_A)
```

# Критерий Лапласса

```python
def laplace(A):
    v = np.mean(A, axis=1)
    return np.argmax(v) + 1
```

```python
laplace(tmp_A)
```

# Критерий Байеса

```python
def bayes(A, p):
    v = [np.sum(p[j]*A[:, j]) for j in range(A.shape[1])]
    return np.argmax(v) + 1
```

```python
bayes(tmp_A, tmp_p)
```

# Критерий Гурвица

```python
def hurwitz(A, lamb):
    v = lamb * np.min(A, axis=1) + (1 - lamb) * np.max(A, axis=1)
    return np.argmax(v) + 1
```

```python
hurwitz(tmp_A, tmp_lamb)
```

# Критерий Сэвиджа

```python
def savage(A):
    r = np.max(A, axis=0) - A
    v = - np.max(r, axis=1)
    return np.argmax(v) + 1    
```

```python
savage(tmp_A)
```
